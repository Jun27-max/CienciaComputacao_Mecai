```python
import random
import string
import pandas as pd

# ------------------------------
# Geração de chaves de cliente únicas
def gerar_chave_cliente():
    return "C" + ''.join(random.choices(string.digits, k=3))

# ------------------------------
# Gerador de produtos simulados
def gerar_dados_entrada(qtd_produtos=10):
    dados = []
    for _ in range(qtd_produtos):
        chave = gerar_chave_cliente()
        tipo = random.choice([1, 2])
        tamanho = random.randint(1, 100)
        dados.append({
            "chave_cliente": chave,
            "tipo_produto": tipo,
            "tamanho": tamanho
        })
    return dados

# ------------------------------
# Interface de recepção dos dados pelo módulo de inspeção
def receber_dados_inspecao(dados_entrada):
    print(f"\nRecebendo {len(dados_entrada)} produtos para inspeção...\n")
    for i, produto in enumerate(dados_entrada, 1):
        print(f"[{i}] Cliente: {produto['chave_cliente']}, Tipo: {produto['tipo_produto']}, Tamanho: {produto['tamanho']}")
    return dados_entrada

# ------------------------------
# Interface de exibição dos dados recebidos
def exibir_lista_dados(dados_entrada):
    df = pd.DataFrame(dados_entrada)
    print("\nLista de produtos fornecidos ao sistema:\n")
    print(df.to_string(index=False))

# ------------------------------
# Exemplo de execução da simulação
if __name__ == "__main__":
    dados = gerar_dados_entrada(qtd_produtos=10)
    produtos_recebidos = receber_dados_inspecao(dados)
    exibir_lista_dados(produtos_recebidos)
```
